## **Análisis de Datos**



### **1. Ingreso total por cada tienda**

#### Calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos

```python
# Lista de todas las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Calcular y mostrar los ingresos
ingresos = []  # Lista para guardar los ingresos totales
for i, t in enumerate(tiendas, start=1):
    ingreso_total = t['Precio'].sum()
    ingresos.append(ingreso_total)
    print(f"Ingreso total de la Tienda {i}: {ingreso_total}")
```

### **2. Ventas por categoría**

#### Calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda

```python
# Lista con los DataFrames de cada tienda
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Calcular productos vendidos por categoría en cada tienda
resultados = {}

for nombre, df in zip(nombres_tiendas, tiendas):
    resumen = (
        df.groupby('Categoría del Producto')['Producto']
        .count()
        .sort_values(ascending=False)
    )
    resultados[nombre] = resumen

# Mostrar resultados
for nombre, resumen in resultados.items():
    print(f"\nCategorías más populares en {nombre}:")
    print(resumen)
```

### **3. Valoración media por tienda**


#### Calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos

```python
# Lista con las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Calcular el promedio de calificación por tienda
promedios = [t['Calificación'].mean() for t in tiendas]

# Nombres de las tiendas
nombres_tiendas = [f"Tienda {i+1}" for i in range(len(tiendas))]

# Mostrar resultados en consola
print("Calificación promedio de los clientes por tienda:")
for i, promedio in enumerate(promedios, start=1):
    print(f"Tienda {i}: {promedio:.2f}")
```

### **4. Productos más y menos vendidos**

#### Identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda

```python
# Lista o diccionario con las tiendas y sus nombres
tiendas = {"Tienda 1": tienda,"Tienda 2": tienda2,"Tienda 3": tienda3,"Tienda 4": tienda4}

# Función para calcular los productos más y menos vendidos
def contar_ventas_por_producto(df):
    return df.groupby('Producto')['Producto'].count().sort_values(ascending=False)

# Función para mostrar los resultados
def mostrar_resultados(nombre_tienda, productos):
    print(f"\n--- {nombre_tienda} ---")
    print(f"Productos más vendidos:\n{productos.head(1)}\n")
    print(f"Productos menos vendidos:\n{productos.tail(1)}\n")

# Iterar sobre todas las tiendas
for nombre, df in tiendas.items():
    productos = contar_ventas_por_producto(df)
    mostrar_resultados(nombre, productos)
```

### **5. Valor del envío promedio por tienda**

#### Calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío

```python
promedio_envio_tienda1 = tienda['Costo de envío'].mean()
promedio_envio_tienda2 = tienda2['Costo de envío'].mean()
promedio_envio_tienda3 = tienda3['Costo de envío'].mean()
promedio_envio_tienda4 = tienda4['Costo de envío'].mean()

print("Costo de envío promedio pagado por los clientes:")
print(f"Tienda 1: {promedio_envio_tienda1:.2f}")
print(f"Tienda 2: {promedio_envio_tienda2:.2f}")
print(f"Tienda 3: {promedio_envio_tienda3:.2f}")
print(f"Tienda 4: {promedio_envio_tienda4:.2f}")
```
